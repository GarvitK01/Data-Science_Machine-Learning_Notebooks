```python
def cleanData(string):
    
    '''
    Tokenizes (Splits into words) and cleans the data (Removes twitter handles, punctuation marks and stop words)
    '''
    
    tknzr = TweetTokenizer(strip_handles=True, reduce_len=True, preserve_case = False)
    string = tknzr.tokenize(string)
    result = ""
    for word in string:  # To convert text into one full string
        result += word + " "
    result = nltk.word_tokenize(result)
    string = [word for word in result if word.isalnum() and word not in stop_words]   #Removing stop_words from the text
    result = ""
    for word in string:
        result += word + " "  #Converting to one whole string again

    return result
```

```python
def makeWordClouds(tag):
    
    '''
    Uses the content of the articles to make WordCloud of all the wods in the REAL/FAKE dataset
    Arguemnts
        tag - "REAL" or "FAKE"
    '''
    
    data = db[db['label'] == tag]
    all_text = ' '.join([text for text in data.text]) #Joins all the words together from the 'text' of the data
    
    wordcloud = WordCloud(width= 800, height= 500,
                          max_font_size = 110,
                          collocations = False).generate(all_text)
    
    plt.figure(figsize=(10,7))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")
    plt.savefig(tag+".jpg")
    plt.show()
```

```python
makeWordClouds("REAL")
makeWordClouds("FAKE")
```

```python
def makeCountPlots(tag, quantity):
    
    data = db[db['label'] == tag]
    all_words = ' '.join([text for text in data.text])
    list_words = nltk.word_tokenize(all_words)
    frequency = nltk.FreqDist(list_words)

    df_frequency = pd.DataFrame({"Word": list(frequency.keys()),
                                       "Frequency": list(frequency.values())})
    df_frequency = df_frequency.nlargest(columns = "Frequency", n = quantity)
    plt.figure(figsize=(12,8))
    ax = sns.barplot(data = df_frequency, x = "Word", y = "Frequency", palette = 'icefire')
    ax.set(ylabel = "Count")
    ax.set(xlabel = 'Words')
    plt.title(tag)
    plt.xticks(rotation='vertical')
    plt.savefig(tag + "freq.jpg")
    plt.show()
```

```python
makeCountPlots("REAL", 10)
makeCountPlots("FAKE", 10)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import TfidfTransformer, CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, plot_confusion_matrix, classification_report
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
def create_dataset():
    
    X = db.text
    y = db['label'].apply(lambda X: 1 if X=="REAL" else 0)
    
    
    X_train, X_test, y_train, y_test = train_test_split(db['text'], y, test_size = 0.2, shuffle = True)
    
    return X_train, X_test, y_train, y_test
```

```python
X_train, X_test, y_train, y_test = create_dataset()
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
pipe = Pipeline([('vect', CountVectorizer()),
                 ('tfidf', TfidfTransformer()),
                 ('model', LogisticRegression())])

model = pipe.fit(X_train, y_train)
prediction = model.predict(X_test)
print("accuracy: {}%".format(accuracy_score(y_test, prediction)*100))
plot_confusion_matrix(model, X_test, y_test, cmap = 'viridis')
plt.show()
print(classification_report(y_test, prediction))
```

```python
pipe = Pipeline([('vect', CountVectorizer()),
                 ('tfidf', TfidfTransformer()),
                 ('model', MultinomialNB())])

model = pipe.fit(X_train, y_train)
prediction = model.predict(X_test)
print("accuracy: {}%".format(accuracy_score(y_test, prediction)*100))
plot_confusion_matrix(model, X_test, y_test, cmap = 'Reds')
plt.show()
print(classification_report(y_test, prediction))
```

```python
pipe = Pipeline([('vect', CountVectorizer()),
                 ('tfidf', TfidfTransformer()),
                 ('model', RandomForestClassifier())])

model = pipe.fit(X_train, y_train)
prediction = model.predict(X_test)
print("accuracy: {}%".format(accuracy_score(y_test, prediction)*100))
plot_confusion_matrix(model, X_test, y_test, cmap = 'RdPu')
plt.show()
print(classification_report(y_test, prediction))
```

```python
pipe = Pipeline([('vect', CountVectorizer()),
                 ('tfidf', TfidfTransformer()),
                 ('model', KNeighborsClassifier( n_neighbors = 3))])

model = pipe.fit(X_train, y_train)
prediction = model.predict(X_test)
print("accuracy: {}%".format(accuracy_score(y_test, prediction)*100))
plot_confusion_matrix(model, X_test, y_test, cmap = 'RdPu')
plt.show()
print(classification_report(y_test, prediction))
```
